# 3 Data preprocessing

```python
A.isna().sum()
```

```python
import missingno as mno
```

```python
mno.bar(A)#Finding out missing value in dataset in graphical form using missingno library
```

```python
A.isna().sum().plot(kind="barh")#Finding out missing value in dataset in graphical form
```

```python
#Treating missing values of continuous columns
```

```python
x = A[["Rear.seat.room"]].mean()
```

```python
x[0]
```

```python
A['Rear.seat.room'] = A['Rear.seat.room'].fillna(x[0])
```

```python
x = A[["Luggage.room"]].mean()[0]
A['Luggage.room']=A['Luggage.room'].fillna(x)
```

```python

#Treating missing values of categorical columns
```

```python
A.AirBags.value_counts()
```

```python
A.AirBags.mode()
```

```python
x = A.AirBags.mode()[0]
A.AirBags = A.AirBags.fillna(x)
```
