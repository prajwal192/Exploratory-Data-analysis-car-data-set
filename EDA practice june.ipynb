```python
import pandas as pd
```

```python
A=pd.read_csv("E:\data science class/50_Startups.csv")
```

```python
A.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,4,1)
plt.scatter(A.RND,A.MKT)
plt.xlabel("RND")
plt.ylabel("MKT")
plt.title("scatter plot bet RND and MKT")


plt.subplot(2,4,2)
plt.scatter(A.MKT,A.PROFIT)
plt.xlabel("MKT")
plt.ylabel("PROFIT")
plt.title("scatter plot bet MKT and PROFIT")

plt.subplot(2,4,3)
plt.scatter(A.RND,A.PROFIT)
plt.xlabel("RND")
plt.ylabel("PROFIT")
plt.title("scatter plot bet RND and PROFIT")

plt.subplot(2,4,5)
sb.distplot(A.RND)
plt.title("DIST PLOT OF RND")

plt.subplot(2,4,6)
A.PROFIT.hist()
plt.title("hist of profit")

plt.subplot(2,4,7)
A.STATE.value_counts().plot(kind="barh")
plt.title("value count of states")

plt.subplot(2,4,8)
sb.distplot(A.PROFIT)
plt.title("dist plot of profit")

plt.savefig("C:/Users/BHUSHAN/Desktop.png",format="png")











```
